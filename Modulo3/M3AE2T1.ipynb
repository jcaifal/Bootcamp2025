```python
import pandas as pd

#ventas = pd.Series([1500, 2300, 1800, 2700, 3000], index=['Ene', 'Feb', 'Mar', 'Abr', 'May'])


#datos = [['Ana', 28, 'Santiago'],['Luis', 34, 'Valparaíso'], ['Marta', 22, 'Concepción']]

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'

df = pd.read_csv(url)

print(df.head())  # Muestra las primeras filas del DataFrame


```

```python
data = {'Nombre': ['Ana', 'Luis', 'Marta'],
        'Edad': [28, 34, 22],
        'Ciudad': ['Santiago', 'Valparaíso', 'Concepción']}

df2 = pd.DataFrame(data)
print(df2)  # Muestra el DataFrame creado a partir del diccionario


nombres = df2['Nombre']
print(nombres)  # Muestra la columna 'Nombre' del DataFrame df2

fila = df2.iloc[0]  # Selecciona la segunda fila del DataFrame df2
print(fila)  # Muestra la primera fila del DataFrame df2

jovenes = df2[df2['Edad'] < 25  ]  # Filtra el DataFrame df2 para mostrar solo las filas donde la edad es menor a 30
print(jovenes)  # Muestra el DataFrame filtrado con jóvenes menores de 25 años

df2['Año_Nacimiento'] = 2025 - df2['Edad']  # Agrega una nueva columna 'Año_Nacimiento' al DataFrame df2
df2['Es_Mayor'] = df2['Edad'] >= 18  # Agrega una nueva columna 'Es_Mayor' al DataFrame df2 
print(df2)  # Muestra el DataFrame df2 con las nuevas columnas añadidas
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 5))  # Establece el tamaño de la figura

for especie in df['species'].unique():
    subset = df[df['species'] == especie]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=especie)

plt.title('Longitud vs Ancho del Sépalo por Especie')
plt.xlabel('Longitud del Sépalo (cm)')  
plt.ylabel('Ancho del Sépalo (cm)')
plt.legend(title='Especie')
plt.grid(True)  # Añade una cuadrícula al gráfico
plt.show()  # Muestra el gráfico
```

```python
ventas = pd.Series([1200, 1350, 1600], index=['Enero', 'Febrero', 'Marzo'])
#imprime las ventas
print(ventas)  # Muestra las ventas de cada mes

#calcule el promedio de las ventas
promedio_ventas = ventas.mean()  # Calcula el promedio de las ventas
print(f"El promedio de las ventas es: {promedio_ventas}")  # Muestra el promedio de las ventas

#acceder al valor de ventas de febrero
ventas_febrero = ventas['Febrero']  # Accede al valor de ventas de febrero
print(f"Las ventas de febrero son: {ventas_febrero}")  # Muestra las ventas de febrero


#determine el mes con mayores ventas
mes_max_ventas = ventas.idxmax()  # Encuentra el mes con mayores ventas 
print(f"El mes con mayores ventas es: {mes_max_ventas}")  # Muestra el mes con mayores ventas

```

```python


df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

print(df.head(5))  # Muestra las primeras filas del DataFrame

df.info()  # Muestra información general del DataFrame, incluyendo el número de filas, columnas y tipos de datos

df.describe()  # Muestra estadísticas descriptivas del DataFrame, como la media, desviación estándar, etc.

#detectar valores nulos
print(f"Valores Nulos: {df.isnull().sum()}")  # Muestra el número de valores nulos en cada columna del DataFrame    

#obtener la media   
print(f"Media: {df['age'].mean()}")  # Muestra la media de la columna 'age' del DataFrame

#obtener la mediana
print(f"Mediana: {df['age'].median()}")  # Muestra la mediana de la columna 'age' del DataFrame

#obtener la desviación estándar
print(f"Desviación Estándar: {df['age'].std()}")  # Muestra la desviación estándar de la columna 'age' del DataFrame

```
