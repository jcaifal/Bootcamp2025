```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ===== 1. Cargar dataset =====
df = pd.read_csv('datos_regresion_1.csv', sep=';', encoding='latin1')

print("\n Resumen inicial:")
print(df.info())
print(df.describe(include='all'))
```

```python

# Suponiendo que ya cargaste el DataFrame como df
X = df[['Horas_Estudio']].values  # Variable independiente (2D array)
y = df['Nota_Final'].values       # Variable dependiente (1D array)


# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Predicciones
y_pred = modelo.predict(X)

# Métricas del modelo
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python

# Imprimir métricas
print(f"Coeficiente de regresión (pendiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")
print(f"Error cuadrático medio (MSE): {mse:.4f}")
print(f"Coeficiente de determinación (R²): {r2:.4f}")
```

```python

# Graficar los datos y la línea de regresión
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Nota Final')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.grid(True)
plt.show()
```
