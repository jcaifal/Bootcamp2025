```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ===== 1. Cargar dataset =====
df = pd.read_csv('ejercicio_2. publicidad_vs_ventas.csv', sep=';', encoding='latin1')

print("\n Resumen inicial:")
print(df.info())
print(df.describe(include='all'))
```

```python


# Reemplazar comas por puntos y convertir a float
df['Inversion_Redes_USD'] = df['Inversion_Redes_USD'].astype(str).str.replace(',', '.').astype(float)
df['Ventas_Miles_Unidades'] = df['Ventas_Miles_Unidades'].astype(str).str.replace(',', '.').astype(float)

# Definir variables
X = df[['Inversion_Redes_USD']].values  # Variable independiente
y = df['Ventas_Miles_Unidades'].values  # Variable dependiente

```

```python


# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)

```

```python

# Realizar predicciones
y_pred = modelo.predict(X)

```

```python

# 5. Evaluar el modelo

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python


# Imprimir métricas
print(f"Coeficiente de regresión (pendiente): {modelo.coef_[0]:.4f}")
print(f"Intercepto: {modelo.intercept_:.4f}")
print(f"Error cuadrático medio (MSE): {mse:.4f}")
print(f"Error absoluto medio (MAE): {mae:.4f}")
print(f"Coeficiente de determinación (R²): {r2:.4f}")

```

```python


# Graficar los resultados
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.xlabel('Inversión en Redes (USD)')
plt.ylabel('Ventas (Miles de Unidades)')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.grid(True)
plt.show()

```

Conclusiones del modelo:
Relación positiva clara:

A medida que aumenta la inversión en redes, también aumentan las ventas.
La línea de regresión roja muestra una tendencia ascendente, lo que indica una correlación positiva.
Buen ajuste del modelo:

Si el coeficiente de determinación R² es alto (como se mostró antes: 0.9955), significa que el modelo explica muy bien la variabilidad de las ventas.
Los puntos están cerca de la línea, lo que sugiere bajo error de predicción.
Modelo adecuado:

La regresión lineal parece suficiente para este conjunto de datos.
No se observan patrones curvos ni dispersión que sugieran la necesidad de un modelo más complejo (como polinómico o árboles).
Aplicabilidad práctica:

Este modelo puede ser útil para estimar ventas futuras en función de la inversión.
También puede ayudar en la toma de decisiones presupuestarias para campañas de redes.
