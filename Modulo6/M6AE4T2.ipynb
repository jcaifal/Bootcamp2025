```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression




# ===== 1. Cargar dataset =====
df = pd.read_csv('datos_regresion_2.csv', sep=';', encoding='latin1')

print("\n Resumen inicial:")
print(df.info())
print(df.describe(include='all'))
```

```python
# Suponiendo que ya cargaste el DataFrame como df
X = df[['Variable_X']].values  # Variable independiente (2D array)
Y = df['Variable_Y'].values       # Variable dependiente (1D array)



# Transformación polinómica (grado 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajustar el modelo
model = LinearRegression()
model.fit(X_poly, Y)

# Predicciones
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
Y_pred = model.predict(poly.transform(X_range))
```

```python
# Graficar
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='green', label='Datos reales')
plt.plot(X_range, Y_pred, color='red', label='Modelo Polinómico')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Regresión Polinómica (Grado 2)')
plt.legend()
plt.grid(True)
plt.show()

```
